```python
# 使用Q-learning算法去迭代
def get_state(state):
    for i in range(10):
        for j in range(10):
            if state[i, j] == 1:
                return 10 * i + j


class SolutionFormat:
    """The code that is asked to fill"""
    def __init__(self):
        """Any initialisation that you want. 
           If you use an algorithm that uses a Q-table (as advised), you can initialize your Q-table here
           You can set any hyperparameter here.
        """
        self.q_table = None 
        self.epsilon = 0.9
        self.gamma = 0.99
        self.lr = 0.01

    def action(self, state):
        """Return an action to do."""
        state = get_state(state)
        # 采用\epsilon-greedy策略
        if np.random.random() > self.epsilon or sum(self.q_table[state]) == 0:
            idx = np.random.randint(4)
        else:
            idx = np.argmax(self.q_table[state])
        return idx
    
    def update(self, state, action ,new_state, reward, done, info, episode):
        """Update the policy."""
        state = get_state(state)
        q = self.q_table[get_state(new_state)]
        q_target = self.q_table[state,action]
        if done:
            q_predict = reward
        else:
            q_predict = reward + self.gamma * np.max(q)
        
        self.q_table[state, action] += self.lr * (q_predict - q_target)

    def reset(self) : 
        """Reset your algorithm to be adapted to other maze. 
           For instance, it could be to reset to 0 the values of the Q-tables
        """
        np.random.seed(2024)
        self.q_table = np.zeros((100,4))
```

```python
# test_code
solution = SolutionFormat()
solution.reset()
n_episodes = 100
for e in range(n_episodes):
    state = env.reset()
    rewards = 0
    done = False
    while not done:
        action = solution.action(state)
        new_state, reward, done, info = env.step(action)
        solution.update(state, action, new_state, reward, done, info, e)
        state = new_state
        rewards += reward
```

```python
solution = SolutionFormat()
solution.reset()
env.reset()
evaluation = EvalRL(env=env,  solution=solution)
score = evaluation.eval()
print(f'Score: {score}')
```
